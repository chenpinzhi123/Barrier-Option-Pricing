```python
import pandas as pd
import numpy as np
from scipy.stats import norm
import copy
from Options.Classes import Vanilla, Barrier
from matplotlib import pyplot as plt

%matplotlib inline
```

```python
_k = 100
_r = 0.025
_q = 0.03
_sigma = 0.25
_t = 40/252

_s = np.arange(75, 125, 0.01)

_typeflag = "c"
_hu = 115
_hd = 95
```

# Prices

```python
for barrier in ("do","di","uo","ui"):
    
    if barrier[0] == "u":
        _h = _hu
    elif barrier[0] == "d":
        _h = _hd

    _rebate = 0

    c = Vanilla(_s, _k, _r, _q, _sigma, _t, _typeflag)
    c_v = c.valuation(s=_s)

    b = Barrier(_s, _k, _r, _q, _sigma, _t, _h, _rebate, barrier, _typeflag)
    c_b_v = b.valuation(s=_s)

    plt.figure()
    plt.plot(_s, c_v, label="Vanilla")
    
    if barrier[0] == "u":
        plt.plot(_s, c_b_v, label="%s Barrier"%barrier)
    elif barrier[0] == "d":    
        plt.plot(_s, c_b_v, label="%s Barrier"%barrier)
        
    plt.legend()
    plt.show()
```
